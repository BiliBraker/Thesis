```python
import pandas as pd 
path = r'c:\Users\soirk\Krisztian\Egyetem\Survey Statisztika Msc\Szakdolgozat\corpus\\'

papers_imput=pd.read_excel(path+'papers_imputation.xlsx',encoding='UTF-8')

papers_imput['advanced']=''
papers_imput['basic']=''
```

```python
# search a keyword in a 20 wide window
# imputation 
def imputation_classifier(keywords,data_frame,context_col_name,class_col_name):
    """
    Takes the keywords and checks whether they are near to the keyword "imput".
    -keywords: keywords to check
    -data_frame: a pandas data frame containing the context and class column
    -context_col_name: name of the context column (ex.:df['name']) 
    -class_col_name: name of the classification column (the values are going to be 0-s and 1-s)
    """
    import pandas as pd
    import nltk
    tokenizer = nltk.RegexpTokenizer(r"\w+")
    n=0
    df=data_frame
    context_col=context_col_name
    class_col=class_col_name
    for j in range(len(keywords)):
        for i in range(len(df[context_col])):
            n+=1
            num=int(n/len(keywords))
            if len(df[context_col])>9:
                context = df[context_col][i]
                context_token=tokenizer.tokenize(context.lower())
                context_clean=[word for word in context_token if word.isalpha() and len(word)>1]
                for w in range(len(context_clean)):
                    if 'imput' in context_clean[w] and w>9 and w<len(context_clean)-10:
                        window=context_clean[w-10:w+10]
                        if keywords[j] in str(window):
                            print(keywords[j],num)
                            df[class_col][i]=1
                    elif 'imput' in context_clean[w] and w<10:
                        window=context_clean[w-(10-w):w+10]
                        if keywords[j] in str(window):
                            print(keywords[j],num)
                            df[class_col][i]=1
                    elif 'imput' in context_clean[w] and w>len(context_clean)-10:
                        window=context_clean[w-10:w+(len(context_clean)-w)]
                        if keywords[j] in str(window):
                            print(keywords[j],num)
                            df[class_col][i]=1
                    else:
                        continue
            else:
                continue
```

```python
# example of tokenizations
import nltk
tokenizer = nltk.RegexpTokenizer(r"\w+")
print(tokenizer.tokenize("em-algorithm"))
print(nltk.word_tokenize("em-algorithm"))
```

```python
keywords=['multiple','algorithm','fiml']

imputation_classifier(keywords,papers_imput,class_col_name='advanced',context_col_name='imput_location')
```

```python
keywords=['mean','median','mode','neighbor','deck']

imputation_classifier(keywords,papers_imput,class_col_name='basic',context_col_name='imput_location')
```
