```python
# arxiv scrape

import urllib
import requests
from bs4 import BeautifulSoup
import pandas as pd



url = 'https://arxiv.org/search/advanced?advanced=&terms-0-operator=AND&terms-0-term=%22missing+data%22&terms-0-field=all&classification-physics_archives=all&classification-include_cross_list=include&date-filter_by=all_dates&date-year=&date-from_date=&date-to_date=&date-date_type=submitted_date&abstracts=show&size=200&order=announced_date_first'



page = requests.get(url)

soup = BeautifulSoup(page.content, 'html.parser')

for link in soup.find_all('a'):
    print(link.get('href'))
```

```python
papers = soup.find_all('p', class_='title is-5 mathjax')
links = soup.find_all('', class_='list-title is-inline-block')


papers_list = []
links_list = []


for i in papers:
    #print(i.get_text())
    papers_list.append(str(i.get_text()).replace('\n',''))
    
for j in links:
    #print(j)
    children = j.findChildren("a" , recursive=False)
    for child in children:
        #print(child.get('href'))
        links_list.append(child.get('href'))
```

```python
papers_list = []
links_list = []


for page_num in range(0,801,200):
    
    url = ('https://arxiv.org/search/advanced?advanced=&terms-0-operator=AND&terms-0-term='+
        '%22missing+data%22&terms-0-field=all&classification-physics_archives'+
        '=all&classification-include_cross_list=include&date-filter_by='+
        'all_dates&date-year=&date-from_date=&date-to_date=&date-date_type=submitted_date&abstracts='+
        'show&size=200&order=announced_date_first&start='+str(page_num))
        #print(page_num)
        
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    
    papers = soup.find_all('p', class_='title is-5 mathjax')
    links = soup.find_all('', class_='list-title is-inline-block')
    
    for i in papers:
        #print(i.get_text())
        papers_list.append(str(i.get_text()).replace('\n',''))
    
    for j in links:
        #print(j)
        children = j.findChildren("a" , recursive=False)
        for child in children:
            #print(child.get('href'))
            links_list.append(str(child.get('href')).replace('abs','pdf'))
            #print(len(links_list))
    
```
