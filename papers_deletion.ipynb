```python
# search a keyword in a 20 wide window
# deletion 
def deletion_classifier(target,keywords,data_frame,context_col_name,class_col_name):
    """
    Takes the keywords and checks whether they are near to the specified keyword.
    -target: a keyword in the center of the window (str)
    -keywords: keywords to check in the window (list)
    -data_frame: a pandas data frame containing the context and class column (pd.DataFrame)
    -context_col_name: name of the context column (ex.:df['name']) 
    -class_col_name: name of the classification column (the values are going to be 0-s and 1-s)
    """

    import pandas as pd
    import nltk
    tokenizer = nltk.RegexpTokenizer(r"\w+")
    # parameters
    n=0
    df=data_frame
    context_col=context_col_name
    class_col=class_col_name
    df[class_col]=''
    for j in range(len(keywords)):
        for i in range(len(df[context_col])):
            n+=1
            num=int(n/len(keywords))
            if len(df[context_col])>24:
                context = df[context_col][i]
                context_token=tokenizer.tokenize(context.lower())
                context_clean=[word for word in context_token if word.isalpha() and len(word)>1]
                for w in range(len(context_clean)):
                    if target in context_clean[w] and w>9 and w<len(context_clean)-10:
                        window=context_clean[w-10:w+10]
                        if keywords[j] in str(window):
                            print(keywords[j],num)
                            df[class_col][i]=1
                    elif target in context_clean[w] and w<10:
                        window=context_clean[w-(10-w):w+10]
                        if keywords[j] in str(window):
                            print(keywords[j],num)
                            df[class_col][i]=1
                    elif target in context_clean[w] and w>len(context_clean)-10:
                        window=context_clean[w-10:w+(len(context_clean)-w)]
                        if keywords[j] in str(window):
                            print(keywords[j],num)
                            df[class_col][i]=1
                    else:
                        continue
            else:
                continue
```

```python
deletion_classifier('cases',['exclud','omit','discard'],papers_del,'deletion_concat2','deletion_class2')
```
