```python
import pandas as pd
import os
import numpy as np

path = r'c:\Users\soirk\Krisztian\Egyetem\Survey Statisztika Msc\Szakdolgozat\corpus\\'

papers_id=pd.read_excel(path+'papers_keywords.xlsx',encoding='UTF-8')
```

```python
papers_id=papers_id.drop(['id'], axis=1)
papers_id.rename(columns={'Column3':'id'}, inplace=True)
```

```python
import nltk
from nltk import bigrams, trigrams
import pandas as pd
import os
from nltk import word_tokenize
from nltk.collocations import *
import enchant

d = enchant.Dict("en_US")
tokenizer = nltk.RegexpTokenizer(r"\w+")

n=0
path = r'c:\Users\soirk\Krisztian\Egyetem\Survey Statisztika Msc\Szakdolgozat\corpus\all_paper_txt'+'\\'
#i=0

imput_bigrams=[]
for i in range(len(papers_id['id'])):
    if (papers_id.iloc[i,2] > 0 and papers_id.iloc[i,2] < 6) and ((papers_id.iloc[i,16] > 0 and papers_id.iloc[i,16] < 2) or (papers_id.iloc[i,2] > 0 and papers_id.iloc[i,2] < 6)):
                paper = open(path+papers_id['id'][i],encoding='utf-8')
                paper_token=tokenizer.tokenize(paper.read().lower())
                paper_clean=[word for word in paper_token if word.isalpha()] #and d.check(word) == True]
                bigram_coll=BigramCollocationFinder.from_words(paper_clean)
                paper_bigram=list(bigram_coll.ngram_fd.items())
                for bi in paper_bigram:
                    if 'imput' in str(bi) or 'substit' in str(bi):
                        imput_bigrams.append(bi)
                        
                papers_id.iloc[i,-1]=str(imput_bigrams)
                imput_bigrams=[]
    print(n)
    n+=1
```

```python
#search for deletion bigrams

import nltk
from nltk import bigrams, trigrams
import pandas as pd
import os
from nltk import word_tokenize
from nltk.collocations import *
import enchant

d = enchant.Dict("en_US")
tokenizer = nltk.RegexpTokenizer(r"\w+")

n=0
path = r'c:\Users\soirk\Krisztian\Egyetem\Survey Statisztika Msc\Szakdolgozat\corpus\all_paper_txt'+'\\'
#i=0

delet_bigrams=[]
for i in range(len(papers_id['id'])):
    if ((papers_id['delet'][i] > 0 and papers_id['delet'][i] < 6) or (papers_id['case'][i] > 0 
    and papers_id['case'][i] < 6) or (papers_id['listwise'][i] > 0)):
                paper = open(path+papers_id['id'][i],encoding='utf-8')
                paper_token=tokenizer.tokenize(paper.read().lower())
                paper_clean=[word for word in paper_token if word.isalpha()] #and d.check(word) == True]
                bigram_coll=BigramCollocationFinder.from_words(paper_clean)
                paper_bigram=list(bigram_coll.ngram_fd.items())
                for bi in paper_bigram:
                    if 'delet' in str(bi) or 'list' in str(bi) or 'pair' in str(bi) or 'exclud' in str(bi) or 'discard' in str(bi) or 'eliminat' in str(bi):
                        delet_bigrams.append(bi)
                        
                papers_id.iloc[i,-1]=str(delet_bigrams)
                delet_bigrams=[]
    print(n)
    n+=1
```
