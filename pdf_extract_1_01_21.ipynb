```python
# extracting text from one paper

Path = "c:/Users/soirk/Krisztian/Egyetem/Survey Statisztika Msc/Szakdolgozat/pdfs/"
filelist = os.listdir(Path)

pages = []
text = []

def text_extractor(path):
    with open(path, 'rb') as f:
        pdf = PdfFileReader(f)
 
        # get pages
        for i in range(0,6):
            pages.append(pdf.getPage(i))
            #print(pages[i])
            #print('Page type: {}'.format(str(type(pages[i]))))
 
            text.append(str(pages[i].extractText()))
            #print(text)
 
 
if __name__ == '__main__':
    path = Path + filelist[0]
    text_extractor(path)
    
    
token = word_tokenize(text[0])
token
```
