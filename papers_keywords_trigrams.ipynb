```python
import pandas as pd
import os
import numpy as np

path = r'c:\Users\soirk\Krisztian\Egyetem\Survey Statisztika Msc\Szakdolgozat\corpus\\'

papers_id=pd.read_excel(path+'papers_keywords.xlsx',encoding='UTF-8')

selected=np.array([1,9,10,24])

papers_id=papers_id.iloc[:,selected]
papers_id.rename(columns={'Column3':'id'}, inplace=True)

papers_id['imput_tri_grams']=''
papers_id.iloc[:,-1]
```

```python
import nltk
from nltk import bigrams, trigrams
import pandas as pd
import os
from nltk import word_tokenize
from nltk.collocations import *
import enchant

d = enchant.Dict("en_US")
tokenizer = nltk.RegexpTokenizer(r"\w+")

n=0
path = r'c:\Users\soirk\Krisztian\Egyetem\Survey Statisztika Msc\Szakdolgozat\corpus\all_paper_txt'+'\\'
#i=0

imput_trigrams=[]
for i in range(len(papers_id['id'])):
    if (papers_id.iloc[i,2] > 0 and papers_id.iloc[i,2] < 6) or ((papers_id.iloc[i,3] > 0 and papers_id.iloc[i,3] < 2) and (papers_id.iloc[i,2] > 0 and papers_id.iloc[i,2] < 6)):
                paper = open(path+papers_id['id'][i],encoding='utf-8')
                paper_token=tokenizer.tokenize(paper.read().lower())
                paper_clean=[word for word in paper_token if word.isalpha()] #and d.check(word) == True]
                trigram_coll=TrigramCollocationFinder.from_words(paper_clean)
                paper_trigram=list(trigram_coll.ngram_fd.items())
                for tri in paper_trigram:
                    if 'imput' in str(tri) or 'substit' in str(tri):
                        imput_trigrams.append(tri)
                        
                papers_id.iloc[i,-1]=str(imput_trigrams)
                imput_trigrams=[]
    print(n)
    n+=1
```
